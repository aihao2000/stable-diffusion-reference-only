```python
import torch
```

```python
n=77
m=27
```

```python
text_embeds = torch.zeros((1, 77, 2048))
```

```python
image_embeds=torch.ones((1,2048))
image_embeds.shape
```

```python
image_embeds = image_embeds.repeat(n - m, 1)
image_embeds.shape
```

```python
image_embeds = image_embeds.unsqueeze(0)
```

```python
embeds = torch.concat([text_embeds[:, :m, :], image_embeds], dim=1)
```

```python
embeds.shape
```

```python
tensor_list = embeds.tolist()

# 
for i in tensor_list:
    for row in i:
        for element in row:
            print(element, end=" ")
        print()
```
